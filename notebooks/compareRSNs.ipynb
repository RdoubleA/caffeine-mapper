### Examine induced subgraphs of specific RSNs

Take the original graph and extract only the nodes that have an activation of a particular RSN.

```python
def drawRSN(G,rsn):
    plt.close('all')
    # Pull nodes belonging to the RSN
    subnodes = [node for node in G.nodes() if G.node[node]['group'][rsn]]
    # Extract subgraph
    subG = G.subgraph(subnodes)
    # Draw graph
    nx.draw_spring(subG)
    plt.show()
```

```python
drawRSN(Greals[80],'Somatomotor')
```

```python
drawRSN(Gnulls[80],'Somatomotor')
```
